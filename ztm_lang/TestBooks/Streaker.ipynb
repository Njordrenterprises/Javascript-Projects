```python
import matplotlib.pyplot as plt

# Coordinates
A = (6, 2)
B = (3, 3)
origin = (0, 0)

# Create the plot
plt.figure()

# Plot points A and B
plt.plot(A[0], A[1], 'ro', label='A (6,2)')  # Point A
plt.plot(B[0], B[1], 'bo', label='B (3,3)')  # Point B

# Draw tails from origin to A and B
plt.plot([origin[0], A[0]], [origin[1], A[1]], 'r-')  # Tail to A
plt.plot([origin[0], B[0]], [origin[1], B[1]], 'b-')  # Tail to B

# Labeling
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Plot of Points A and B with Tails to the Origin')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()

```
